```python
import numpy as np 
```

```python
# Load scaler and model
import pickle
scaler = pickle.load(open("scaler.sav", "rb")) 
model = pickle.load(open("rf_model.sav", "rb"))
```

```python
# Example input: age, bmi, children, sex_male, smoker_yes
new_data = np.array([[25, 30.5, 2, 1, 0]])
```

```python
# Scale input
new_data_scaled = scaler.transform(new_data)
```

```python
# Predict
prediction = model.predict(new_data_scaled)
print("Predicted Insurance Cost:", prediction)
```
