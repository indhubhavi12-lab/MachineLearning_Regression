```python
#import pandas library as pd
import pandas as pd
```

```python
#Reading Salary_Data file using read_csv function and assigned to dataset variable
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#split input data
dataset
```

```python
#split input data
independent=dataset[["YearsExperience"]] 
```

```python
#cut the years Experience columnfor the dataset and assigned to independent variable
independent
```

```python
#split output data
dependent=dataset[["Salary"]]
dependent
```

```python
#splitting Traning Set and Test Set using sklearn library
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#learning Phase
from sklearn.linear_model import LinearRegression #import LinearRegression func from sklearn.linear.model
regressor=LinearRegression()#Called Linear unparatized func and assigned to regressor variable
regressor.fit(X_train, Y_train) # Call X_train, Y_ Train values using regressor.fit func
```

```python
weight=regressor.coef_#calculating weight using coef_ method
weight
```

```python
bais=regressor.intercept_ #Calculate bais using intercept_ method
bais
```

```python
#evaluate Phase
y_pred=regressor.predict(X_test) #getting predict value usin test set
```

```python
#calculate r2_score and evaluvate test set model
from sklearn.metrics import r2_score #import r2_score func from sklearn.metrics
r_score=r2_score(Y_test,y_pred) # Evaluate r2 score using actual and predict value
```

```python
r_score
```

```python
#Saving the Final Model
import pickle#import pickle liblrary 
filename="finalized_model_Linear.sav"# save the model as finalized_model_linear.sav file and assigned to filename
```

```python
pickle.dump(regressor,open(filename,'wb'))#save the model
```

```python
#load the model to get final result
loaded_model=pickle.load(open("finalized_model_Linear.sav",'rb'))
result=loaded_model.predict([[13]]) #FFinal result
```
